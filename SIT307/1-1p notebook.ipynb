# Create three list variables named 'StudentName', 'ID', and 'EmailAddress'. Add 5 random values in these lists and print them.

```python
def generateStr(letters, length):
    return "".join(random.choice(letters) for i in range(length))

alphabet = 'abcdefghijklmnopqrstuvwxyz'
numbers = '0123456789'
StudentName = []
ID = []
EmailAddress = []

for i in range(5):
    StudentName.append(generateStr(alphabet, random.choice(range(4,10))))
    ID.append(StudentName[i][:2] + (generateStr(numbers, 6)))
    EmailAddress.append(StudentName[i][:3] + '.' + generateStr(alphabet, 5) + '@' + generateStr(alphabet, 8) + '.com')
    
print(StudentName)
print(ID)
print(EmailAddress)
```

# Create a dictionary variable with three keys, which are named same as the list created in the previous problem. Insert the previously created lists in the dictionary variable and print it.

```python
StudentDict = {}
StudentDict['Name'] = StudentName
StudentDict['ID'] = ID
StudentDict['Email'] = EmailAddress
print(StudentDict)
```

#  Define a numeric variable 'energy' and assign value either 1 or 2. Print "Apple" if the value of energy variable is 1 otherwise print "Banana". What will your program print if energy=10? 

```python
def checkEnergy(energy):
    print(energy)
    if energy == 1:
        print('Apple' + '\n')
    else:
        print('Banana' + '\n')

energy = random.choice([1,2])
checkEnergy(energy)
energy = 10
checkEnergy(energy)
```

#  Read the data stored in the "data.csv" file as a matrix. The row of the matrix represents data point or instance and the column represents the variable or feature. Print the number of data instances and variables present in this dataset. Calculate mean of the second variable and deduct this mean from original values of the variable. Print both original and modified values of the variable.

```python
data = pd.read_csv('data.csv', header=None)
dataFrame = pd.DataFrame(data)
finalArray = np.array(dataFrame.values)

print('number of data point: ' + str(len(finalArray)))
print('number of variable: ' + str(len(finalArray[0])))

totalVar = 0
for i in finalArray:
    print(i[1])
    totalVar += i[1]

avg = totalVar/len(finalArray)

modList = []
for i in finalArray:
    modList.append(i[1] - avg)

print('\n')

for i in modList:
    print(i)
```
